```python
def drawRects(frame, rects, color = 'green'):
    if color == 'green':
        clr = (0,255,0)
    elif color == 'blue':
        clr = (255,0,0)
    else:
        clr = (0,0,255)
        
    for rect in rects:
        x,y,w,h = rect
        cv2.rectangle(frame, (x, y), (x+w, y+h), clr, 2)
```

```python
def Visualize(predictions, groundtruth, time_between_frames = 1000, show_preds = True, show_gts = True):
    os.chdir(main_project_dir)
    
    it = 0
    for im_path, preds in predictions.items():
#         it += 1
#         if it < 350:
#             continue
        
        frame = cv2.imread(im_path)

        gts = []
        try:
            gts = groundtruth[im_path]
        except:
            pass

        table = main_metric(gts, preds, frame.shape)
        
        if show_gts:
            drawRects(frame, gts, 'blue')

        for j, pred in enumerate(preds):
            x,y,w,h = pred
            if table.shape[0] * table.shape[1] != 0 :
                IoUs = table[:,j,0]
                IoOs = table[:,j,1]
                try:
                    is_true = check_true(IoUs, IoOs, "modified")
                except:
                    is_true = False
            else:
                is_true = False
            
            if show_preds:
                if is_true:
                    drawRects(frame, [pred], 'green')
                else:
                    drawRects(frame, [pred], 'red')

        cv2.imshow('frame', frame)

        if cv2.waitKey(time_between_frames) & 0xFF == ord('q'):
            break

    cv2.destroyAllWindows()
    
    os.chdir(predictions_dir)
```

```python
def get_iou(bb1, bb2):
    """
    Calculate the Intersection over Union (IoU) of two bounding boxes.

    Parameters
    ----------
    bb1 : dict
        Keys: {'x1', 'x2', 'y1', 'y2'}
        The (x1, y1) position is at the top left corner,
        the (x2, y2) position is at the bottom right corner
    bb2 : dict
        Keys: {'x1', 'x2', 'y1', 'y2'}
        The (x1, y1) position is at the top left corner,
        the (x2, y2) position is at the bottom right corner

    Returns
    -------
    float
        in [0, 1]
    """
    assert bb1['x1'] < bb1['x2']
    assert bb1['y1'] < bb1['y2']
    assert bb2['x1'] < bb2['x2']
    assert bb2['y1'] < bb2['y2']

    # determine the coordinates of the intersection rectangle
    x_left = max(bb1['x1'], bb2['x1'])
    y_top = max(bb1['y1'], bb2['y1'])
    x_right = min(bb1['x2'], bb2['x2'])
    y_bottom = min(bb1['y2'], bb2['y2'])

    if x_right < x_left or y_bottom < y_top:
        return 0.0

    # The intersection of two axis-aligned bounding boxes is always an
    # axis-aligned bounding box
    intersection_area = (x_right - x_left) * (y_bottom - y_top)

    # compute the area of both AABBs
    bb1_area = (bb1['x2'] - bb1['x1']) * (bb1['y2'] - bb1['y1'])
    bb2_area = (bb2['x2'] - bb2['x1']) * (bb2['y2'] - bb2['y1'])

    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = intersection_area / float(bb1_area + bb2_area - intersection_area)
    assert iou >= 0.0
    assert iou <= 1.0
    return iou

def IOU(cnt1, cnt2):
    x, y, w, h = cnt1
    bb1 = dict(zip(['x1', 'x2', 'y1', 'y2'], [x, x + w, y, y + h]))
    
    x, y, w, h = cnt2
    bb2 = dict(zip(['x1', 'x2', 'y1', 'y2'], [x, x + w, y, y + h]))
    
    return get_iou(bb1, bb2)
```

```python
def tracking(groundtruth, n = 3):
    os.chdir(main_project_dir)
    
    keys = list(groundtruth.keys())
    
    new_keys = keys[n:-n]
    new_groundtruth = {}
    
    for key in new_keys:
        pos = keys.index(key)

        gts = []

        for ith_key in range(pos - n, pos + n, 1):
            gts += groundtruth[keys[ith_key]]
        
        new_groundtruth[key] = gts
    
#     print(new_groundtruth)
    
    for im_path, gts in new_groundtruth.items():
        
        
#         print(im_path, gts)
        frame = cv2.imread(im_path)
        
        groups = []
        
        threshold = 0.3
        
        while len(gts) > 0:
            gt = gts[0]
            
            gt_group = [gt]
            
            for gt2 in gts[1:]:
                if IOU(gt, gt2) > threshold:
                    gt_group.append(gt2)
            
            for gt in gt_group:
                gts.remove(gt)
            
            groups.append(gt_group)
            
            
#         print(groups, len(groups))
            
        for group in groups:
            color = colors[groups.index(group) % len(colors)]
            
            for (x, y, w, h) in group:
                cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)

#         for group in groups:
#             if len(group) > 1:
#                 color = (0,255,0)
#             else:
#                 color = (0,0,255)
            
#             for (x, y, w, h) in group:
#                 cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
        
        cv2.imshow('frame', frame)
        
        init_gts = groundtruth[im_path]
        
        new_gts = set()
        
        for init_gt in init_gts:
            for group in groups:
                if len(group) > 1 and init_gt in group:
                    new_gts.add(tuple(init_gt))
                    
        new_gts = list(new_gts)
        new_gts = list(map(list, new_gts))
        
        groundtruth[im_path] = new_gts

        last_key = cv2.waitKey(100000) & 0xFF
        if last_key == ord('q'):
            break
        else:
            continue
            
        
    cv2.destroyAllWindows()
    
        
    os.chdir(predictions_dir)
```
