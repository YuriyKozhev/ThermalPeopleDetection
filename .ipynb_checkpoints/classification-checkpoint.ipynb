```python
import pandas as pd
import numpy as np
import datetime
import json
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# metric calculating libraries
from new_metric import main_metric, check_true
```

```python
json_folder = "Groundtruth&Predictions(json)//"
gts_folder = json_folder + "Gts//"
preds_folder = json_folder + "Preds//"

VIDEO_train_groundtruth = {}
with open(gts_folder + "VIDEO_train_groundtruth.json", "r") as read_file:
    VIDEO_train_groundtruth = json.load(read_file)
    
VIDEO_val_groundtruth = {}
with open(gts_folder + "VIDEO_val_groundtruth.json", "r") as read_file:
    VIDEO_val_groundtruth = json.load(read_file)
    
FLIR_train_groundtruth = {}
with open(gts_folder + "FLIR_train_groundtruth.json", "r") as read_file:
    FLIR_train_groundtruth = json.load(read_file)
    
FLIR_val_groundtruth = {}
with open(gts_folder + "FLIR_val_groundtruth.json", "r") as read_file:
    FLIR_val_groundtruth = json.load(read_file)
```

```python
VIDEO_train_predictions_TSMO = {}
with open(preds_folder + "VIDEO_train_predictions_TSMO.json", "r") as read_file:
    VIDEO_train_predictions_TSMO = json.load(read_file)
    
FLIR_train_predictions_TSMO = {}
with open(preds_folder + "FLIR_train_predictions_TSMO.json", "r") as read_file:
    FLIR_train_predictions_TSMO = json.load(read_file)
```

```python
metric_type = 'modified'
```

```python
true_false_cnts = {}

# bugs = 0

for im_path in list(VIDEO_train_groundtruth.keys()):
    gts = VIDEO_train_groundtruth[im_path]
    frame = cv2.imread(im_path, 0)

    if im_path in VIDEO_train_predictions_TSMO:
        preds = VIDEO_train_predictions_TSMO[im_path]
    else:
        preds = []
    
    try:
        table = main_metric(gts, preds, frame.shape)

        for j, pred in enumerate(preds):
            x,y,w,h = pred
            pred_int = frame[y:y+h,x:x+w].mean()
            if table.shape[0] * table.shape[1] != 0 :
                IoUs = table[:,j,0]
                IoOs = table[:,j,1]

                max_IoU = IoUs.max()
                max_IoO = IoOs.max()
                is_true = check_true(IoUs, IoOs, metric_type)

                intersected_gts = None
                IoUs = None
            else:
                max_IoU = 0
                max_IoO = None
                intersected_gts = None
                IoUs = None

            if im_path in true_false_cnts:
                true_false_cnts[im_path].append((pred, is_true))
            else:
                true_false_cnts[im_path] = []
                true_false_cnts[im_path].append((pred, is_true))
    except ZeroDivisionError:
        pass
#         print(im_path)
#         bugs += 1

# print(bugs)
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import glob

true = []
false = []

for im_path in list(true_false_cnts.keys()):
    for pair in true_false_cnts[im_path]:
        if pair[1]:
            true.append((im_path, pair[0]))
        else:
            false.append((im_path, pair[0]))
            
np.random.shuffle(true)
np.random.shuffle(false)

print(len(true), len(false))

true_train, true_test = train_test_split(true[:-1], test_size=0.33, random_state=42, shuffle = False)
false_train, false_test = train_test_split(false[:-1], test_size=0.33, random_state=42, shuffle = False)

print(len(true_train), len(false_train))

train = true_train + false_train
#train = np.append(true_train, false_train)
train_labels = np.append(np.ones(len(true_train)), np.zeros(len(false_train)))
train, train_labels = shuffle(train, train_labels, random_state=42)

test = true_test + false_test
#test = np.append(true_test, false_test)
test_labels = np.append(np.ones(len(true_test)), np.zeros(len(false_test)))
test, test_labels = shuffle(test, test_labels, random_state=42)
```
