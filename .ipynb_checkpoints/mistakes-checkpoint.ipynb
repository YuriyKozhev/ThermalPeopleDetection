```python
def mistakes(gr_t, cont):
    
#   gr_t: X,Y,W,H
#   cont: X,Y,X+W,Y+H
    
    bool_gt = [False] * len(gr_t)
    bool_cont = [False] * len(cont)
    bool_gt_check = [True] * len(gr_t)
    
    for i, gr in enumerate(gr_t):
        for j, ct in enumerate(cont):
            if intersecting_detect(gr, ct):
                
#               Пересечение

                left = max(gr[0], ct[0])
                top = max(gr[3] + gr[1], ct[3]);
                right = min(gr[2] + gr[0], ct[2]);
                bottom = min(gr[1], ct[1]);
                
                s1 = (right - left) * (top - bottom)
                
#                Объединение

                x1 = min(gr[0], ct[0])
                y1 = min(gr[1], ct[1])
                x2 = max(gr[2] + gr[0], ct[2])
                y2 = max(gr[3] + gr[1], ct[3])
                
                s2 = (x2 - x1) * (y2 - y1)
                
                if s1 / s2 > .5:
                    bool_gt[i] = True
                    
                    if bool_gt_check[i] == True:
                        bool_cont[j] = True
                        
                    bool_gt_check[i] = False
                    
                    
    fs = bool_gt.count(False)
    ts = bool_gt.count(True)
    
    fc = bool_cont.count(False)
    tc = bool_cont.count(True)
    
    e1 = fc / (fc + tc)
    e2 = fs / (fs + ts)
    
    return e1, e2
```

```python
def intersecting_detect(fir, sec, near = False):

    a = False

    ax1 = fir[0]
    ay1 = fir[1]
    ax2 = fir[2]
    ay2 = fir[3]

    bx1 = sec[0]
    by1 = sec[1]
    bx2 = sec[2]
    by2 = sec[3]

    s1 = ( ax1>=bx1 and ax1<=bx2 ) or ( ax2>=bx1 and ax2<=bx2 )
    s2 = ( ay1>=by1 and ay1<=by2 ) or ( ay2>=by1 and ay2<=by2 )
    s3 = ( bx1>=ax1 and bx1<=ax2 ) or ( bx2>=ax1 and bx2<=ax2 )
    s4 = ( by1>=ay1 and by1<=ay2 ) or ( by2>=ay1 and by2<=ay2 )

    if ((s1 and s2) or (s3 and s4)) or ((s1 and s4) or (s3 and s2)):
        a = True
        return a
        
    elif near:
        if (s1 or s3) and (abs(ay1 - by2) < 35 or abs(by1 - ay2) < 35):
            a = True
            return a
        
        if (abs(ax1 - bx2) < 12 or abs(bx1 - ax2) < 12) and (abs(ay1 - by2) < 22 or abs(by1 - ay2) < 22): 
            a = True
            return a
    else:
        return False
```
